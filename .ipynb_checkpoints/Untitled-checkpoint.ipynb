```python

import geopandas as gp
from geodatasets import get_path
street_population=gp.read_file("data/ludnosc_shp/liczba ludnosci 1_2024_shp/Ludnosc ogolem_2024_01_point.shp")
district_data=gp.read_file("gran_osiedli.geojson")
street_population_geo=street_population.to_crs(epsg=4326)
district_data_geo=district_data.to_crs(epsg=4326)

street_population_proj=street_population.to_crs(epsg=2180)
district_data_proj=district_data.to_crs(epsg=2180)

merged=gp.sjoin(street_population_proj,district_data_proj, how="left", predicate="within")
agregated=merged.groupby("name")["liczba_os"].sum()

result=district_data_proj.merge(agregated, on="name")
result["area"]=result.area/1000000
density=result["liczba_os"]/result["area"]
result["density"]=density
result_converted=result.to_crs(epsg=4326)
result_converted.drop(columns=["id","@id","@relations","addr:postcode","contact:website","name:cs","name:de","teryt:typ","wikidata","wikipedia","name:en","name:prefix","place","population:date"], inplace=True)

```
